```
%scala
import org.apache.spark.sql.functions.col
import org.apache.spark.sql.functions._
import org.apache.spark.sql.types._
import org.apache.spark.sql.Column


val infer_schema = "false"
val first_row_is_header = "false"
val delimiter = " "
val file_location = "/FileStore/tables"
val file_type = "csv"

val df = spark.read.format(file_type)
  .option("inferSchema", infer_schema)
  .option("header", first_row_is_header)
  .option("sep", delimiter)
  .load(file_location)

val b = """\"(GET|POST|HEAD|PUT|DELETE|CONNECT|TRACE|OPTIONS|PATCH)?\s(-|\S+)\s(\S+)\""""
val host="""(\S+)"""
val path = """\s(\S+)"""
val protocol = """\S+\s+\S+\s(\S+).*"""
val a = df.select(regexp_extract($"_c5", host,1))
val z = df.select(regexp_extract($"_c5", path,1))
val primerCorchete = """\["""
val segundoCorchete = """\]"""
//val y = df.select(regexp_extract($"_c5", host,3))
val x = df.select(regexp_extract($"_c5", protocol,1))


val sustituir = df.withColumn("Date",regexp_replace($"_c3",primerCorchete,"")).withColumn("Timezone",regexp_replace($"_c4",segundoCorchete,"")).withColumnRenamed("_c0","Host").withColumnRenamed("_c1","User-identifier").withColumnRenamed("_c2","Userid").withColumn("Request Method",regexp_extract($"_c5",host,1)).withColumn("Resource",regexp_extract($"_c5",path,1)).withColumn("Protocol",regexp_extract($"_c5",protocol,1)).withColumnRenamed("_c6","HTTP status code").withColumnRenamed("_c7","Size")
val data = sustituir.drop("_c3").drop("_c4").drop("_c5")

display(data)
```

```
%scala
val dataF = data.withColumn("Date",to_timestamp(col("Date"), "dd/MMM/yyyy:HH:mm:ss"))
```

```
df.select("Protocol").distinct().show()
```

¿Cuáles son los códigos de estado más comunes en la web? Agrúpalos y ordénalos para ver cuál es el más común.

```
df.groupBy("HTTP status code").count().orderBy("count",ascending=False).show()
```

¿Y los métodos de petición (verbos) más utilizados?

```
df.groupBy("Request Method").count().orderBy("count",ascending=False).show()
```
